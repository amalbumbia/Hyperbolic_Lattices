# Generative Reflection Graph \[Static\] (GRG\[S\]) kernel

In this notebook we provide examples of how tilings made with the generative reflection (GR) kernel can be used

```python
from hypertiling import HyperbolicGraph, GraphKernels
from hypertiling.graphics.plot import plot_tiling
from hypertiling.kernel.GRG_util import plot_graph
import matplotlib.pyplot as plt
```

## Example 1: 3,7-graph

```python
# Generative variant
graph = HyperbolicGraph(3, 7, 5, kernel=GraphKernels.GenerativeReflectionGraph) 
print(f"Graph consists of {len(graph)} nodes")
```

```python
colors = ["#FF000080", "#00FF0080", "#0000FF80"]
plot_graph(graph.get_nbrs_list(), graph.center_coords, graph.p,
                        colors=[colors[graph.get_reflection_level(i) % len(colors)] for i in range(graph.length)])
plt.show()
```

```python
# Static variant
graph = HyperbolicGraph(3, 7, 5, kernel=GraphKernels.GenerativeReflectionGraphStatic) 
print(f"Graph consists of {len(graph)} nodes")
```

```python
colors = ["#FF000080", "#00FF0080", "#0000FF80"]
plot_graph(graph.get_nbrs_list(), graph.center_coords, graph.p,
                        colors=[colors[graph.get_reflection_level(i) % len(colors)] for i in range(graph.length)])
plt.show()
```

As can be seen, the static variant has radially ordered indices as the whole tiling is constructed. The generative variant has radial ordered indices within a slice but globally slicewise ordered indices as only one sector is generated. 

## Example 3: Voter model with boundary condition

In this simple statistic model, cells will always take on the color of the majority of their adjacent cells. As this does not depend on their position (only their neighboring relations, a graph instead of a tiling can be used, too). To highlight the advantages of GRG and GRGS we add some boundary condition. Namely, we want the most outer region to be in a certain state (1) at the start. As the plot function always plots the index too, the plotting itself is a bit slower than for the GR kernel. Therefore, we restrict this example to smaller tilings. 

```python
from hypertiling import HyperbolicGraph, GraphKernels
from hypertiling.graphics.plot import plot_tiling
from hypertiling.kernel.GRG_util import plot_graph

import random
import time
import matplotlib.pyplot as plt
import numpy as np
```

### Generative Case

The usage of the generative Graph has an obvious advantage: Less memory requieremnt. Therefore, larger tilings could be created with the same value of memory available. However, as this is just a small example, we will not exploit this here for time reasons.

```python
t1 = time.time()

p, q, n = 7, 3, 6
graph = HyperbolicGraph(p, q, n, kernel=GraphKernels.GenerativeReflectionGraph) 

print(f"Generation of {len(graph)} polygons took {round(time.time() - t1,5)} s")

# Initialize voter model with random states
states = np.random.randint(0, 2, size=len(graph))

# Set boundary conditions (this is a bit more complicated for generative graphs and radial boundary conditions)
jump = graph._nbrs.shape[0] - 1  # get length of sector
shift = graph._sector_lengths_cumulated[-2]  # get start index in first sector

for i in range(p):  # iterate over all sectors
    start = i * jump + shift
    states[start: start + graph._sector_lengths[-1]] = 1

# plot to show
colors = ["#FFFFFF", "#000000"]
plot_graph(graph.get_nbrs_list(), graph.center_coords, graph.p,
                        colors=[colors[states[i]] for i in range(graph.length)])
plt.show()
```

Run the model and display resulting configuration

```python
its = 1e3 # number of iterations

t1 = time.time()
for i in range(int(its)):
    index = int(len(graph) * np.random.random())
    sum_ = sum([states[nbr] for nbr in graph[index]])
    if sum_ > p // 2:
        states[index] = 1
    else:
        states[index] = 0
print(f"Updateing took {time.time() - t1}s")
    
# plot to show
colors = ["#FFFFFF", "#000000"]
plot_graph(graph.get_nbrs_list(), graph.center_coords, graph.p,
                        colors=[colors[states[i]] for i in range(graph.length)])
plt.show()
```

### Static case

The advantage of the static case is an increased speed for accessing the neighbors. However its disadvantages are a bigger memory requierement as well as a increased construction time.

```python
t1 = time.time()

p, q, n = 7, 3, 6
graph = HyperbolicGraph(p, q, n, kernel=GraphKernels.GenerativeReflectionGraphStatic) 

print(f"Generation of {len(graph)} polygons took {round(time.time() - t1,5)} s")

# Initialize voter model with random states
states = np.random.randint(0, 2, size=len(graph))

# Set boundary conditions (this is a bit more easy for static graphs and radial boundary conditions)
states[graph._sector_lengths_cumulated[-2]:] = 1

# plot to show
colors = ["#FFFFFF", "#000000"]
plot_graph(graph.get_nbrs_list(), graph.center_coords, graph.p,
                        colors=[colors[states[i]] for i in range(graph.length)])
plt.show()
```

```python
its = 1e3 # number of iterations

t1 = time.time()
for i in range(int(its)):
    index = int(len(graph) * np.random.random())
    sum_ = sum([states[nbr] for nbr in graph[index]])
    if sum_ > p // 2:
        states[index] = 1
    else:
        states[index] = 0
print(f"Updateing took {time.time() - t1}s")
    
# plot to show
colors = ["#FFFFFF", "#000000"]
plot_graph(graph.get_nbrs_list(), graph.center_coords, graph.p,
                        colors=[colors[states[i]] for i in range(graph.length)])
plt.show()
```

As can be seen, the GRGS kernel is slower in constructing the tiling but faster in accessing the neighbors. A increased memory footprint is not shown here but to be expected.

## check_integrity
Similar to the GR kernel, GRG(S) provides a method for a short testing of the constructed tiling.

```python
from hypertiling import HyperbolicGraph, GraphKernels
from hypertiling.graphics.plot import plot_tiling
from hypertiling.kernel.GRG_util import plot_graph
import matplotlib.pyplot as plt
```

```python
graph = HyperbolicGraph(5, 4, 5, kernel=GraphKernels.GenerativeReflectionGraph)
graph.check_integrity()
```

This is the expected result. It can be explained by the fact, that the polygon with index 2 is the first polygon of the last layer and thus is missing some neighbors. We can either verify this pretty easily by accessing a protected attribute of GR (fastest and best method), by testing (more intuitive, still fast) or ploting (most intuitive but very slow).

```python
print(f"Expect complete set of neighbors until cell {graph._sector_lengths_cumulated[-2]}")  
```

For demonstration purposes, we access a protected attribute here, `_sector_lengths_cumulated`. If you dont know what protected means: dont touch it! It is an array that stores the start indices of each constructed as well as the next theoretical layer. Therefore, `t._sector_lengths_cumulated[-2]` accesses the first polygon of the last constructed layer.

```python
print(graph.get_reflection_level(12))
```

We can see, that the polygon with index 12 (= 13 - 1) is in the previous layer and thus the polygon with index is the first polygon in the layer with index 6 (layers start counting at 0 and thus in the 5th layer). Therefore, we expect it to miss some of its neighbors causing the message 

```python
fig_ax = plt.subplots(figsize=(8,8))
fig_ax[1].set_xlim(-1, 1)
fig_ax[1].set_ylim(-1, 1)
fig_ax[1].set_box_aspect(1)

colors = ["#FF000080", "#00FF0080", "#0000FF80"]
plot_graph(graph.get_nbrs_list(), graph.center_coords, graph.p,
                        colors=[colors[graph.get_reflection_level(i) % len(colors)] for i in range(graph.length)])
plt.show()
```
