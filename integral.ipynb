```python
import numpy as np

def e2(M, N=1200):
    kx = np.linspace(-np.pi, np.pi, N, endpoint=False) # define integration bounds
    ky = np.linspace(-np.pi, np.pi, N, endpoint=False) # define integration bounds
    dk = (2*np.pi)/N # result of integrating over dky and dkx
    KX, KY = np.meshgrid(kx, ky, indexing='ij') # define possible kx and ky values
    
    sinkx = np.sin(KX)
    sinky = np.sin(KY)
    coskx = np.cos(KX)
    cosky = np.cos(KY)
    f3 = M + 2 - coskx - cosky
    f_abs = np.sqrt(sinkx**2 + sinky**2 + f3**2)
    
    numerator = (M+2)*coskx*cosky - coskx - cosky
    Fz = numerator / (2.0 * f_abs**3) 
    
    integral = np.sum(Fz) * dk * dk
    e2 = (1.0/(2*np.pi)) * integral 
    return e2

for M in [1.0, -1.0]:
    val = e2(M, N=1000)  
    print(f"M={M:+.0f}, e2 ≈ {val:.12f}")
```
